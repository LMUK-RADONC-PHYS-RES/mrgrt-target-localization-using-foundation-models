```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from generate_example_sequence import generate_sequence
```

```python
S, path = generate_sequence(width=100,
                            height=100,
                            num_frames=100,
                            radius=0.5,
                            amplitude=0.4,
                            frequency=2)
```

```python
fig, axs = plt.subplots(ncols=2, sharex=True, sharey=True)

axs[0].plot(path[:,0])
axs[1].plot(path[:,1])

axs[0].set_xlabel('frame')
axs[1].set_xlabel('frame')
axs[0].set_ylabel('x')
axs[1].set_ylabel('y')
axs[0].set_ylim(-1,1)
```

```python
fig, axs = plt.subplots(ncols=2, sharex=True, sharey=True)
axs[0].imshow(S[0], interpolation='nearest')
axs[1].imshow(S[5], interpolation='nearest')
```
